# Welcome !

Welcome to SimpleITK in the IPython Notebook

We start by importing the SimpleITK python module

```python
import SimpleITK as sitk
```

Now we import the modules that will help us download images into the filesystem of the server

```python
import urllib
import os.path

destination = '/tmp/tardigrade001ernesto_moya.jpg'
source = 'http://figshare.com/download/file/1547495'
```

Or we could get the image from a Girder database instance

```python
# destination = '/tmp/tardigrade002.jpg'
# source = 'http://54.221.241.247:8080/api/v1/file/539c7a298d72b60edb223eca/download/IMG_20140331_131725.jpeg'
```

With that, we can pull the image from an online service and copy it as a local file in the server where we are running the IPython notebook

```python
if not os.path.isfile(destination):
    urllib.urlretrieve(source,destination)
```

Now we import the Matplotlib modules that will be used to plot the images in the notebook. Note that the images will be displayed as matrices (their pixel spacing will not be taken into account at this point).

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

We use the ImageFileReader class from SimpleITK to read the image into an sitk.Image object

```python
reader = sitk.ImageFileReader()
reader.SetFileName(destination)
image = reader.Execute()
```

The same can be done with the compacted expression:

```python
image = sitk.ReadImage(destination)
```

We can check some basic properties of the image

```python
print image.GetSize()
```

```python
print image.GetSpacing()
```

```python
print image.GetOrigin()
```

```python
print image.GetDirection()
```

We can use the Matplotlib function "imshow" to display the image. For this, we first extract the pixel data from the image in the form of a numpy array.

```python
imshow(sitk.GetArrayFromImage(image));
```

Note that this is a color image. That is, it has multiple components per pixel. Let's see how many:

```python
print image.GetNumberOfComponentsPerPixel()
```

Therefore, for some of the processing, we want to convert it into an image of Scalar values, to have a single value per pixel. We can do this with the VectorMagnitudeImageFilter. Note that a more formal way to do this would have been to use a computation of Luminance.

```python
#magnitude = sitk.VectorMagnitudeImageFilter()
#mimage = magnitude.Execute(image)
miage = image
```

The same can be done with the more compact expression

```python
#mimage = sitk.VectorMagnitude(image)
mimage=image
```

We can confirm that the image has now a single component

```python
print mimage.GetNumberOfComponentsPerPixel()
```

And we can display this new image

```python
imshow(sitk.GetArrayFromImage(mimage));
```

By default, the imshow() functions use a rainbow colormap. We can replace it with a Grayscale colormap.

```python
imshow(sitk.GetArrayFromImage(mimage),cmap=cm.Greys);
```

Or we could use a Reverse Grayscale colormap: "Greys_r" instead of "Greys".

```python
imshow(sitk.GetArrayFromImage(mimage),cmap=cm.Greys_r);
```

We can find more colormaps at:

http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormap

Now we turn our attention to some basic image processing. Let's threshold the image using the OtsuThresholdImageFilter. This filter computes a threshold value that maximize the separation between two classes based on the intensity distribution in the image.

```python
otsu = sitk.OtsuThresholdImageFilter()
bimage = otsu.Execute(mimage)
```

```python
imshow(sitk.GetArrayFromImage(bimage),cmap=cm.Greys);
```

We can write this file by using the ImageFileWriter class

```python
writer = sitk.ImageFileWriter()
writer.SetFileName('/tmp/ernesto_moyaOtsuThreshold.png')
writer.Execute(bimage);
```

The same can be done with the more compact expression

```python
sitk.WriteImage(bimage,'/tmp/ernesto_moyaOtsuThreshold.png')
```
